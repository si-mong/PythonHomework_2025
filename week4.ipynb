```python
## 수업일: 2025-03-26 수요일
## 3주차과제 2023078086 최은재
## 과제: 5명의 학생의 세개의 교과목 (영어, C-언어, 파이썬)에 대하여 
## 키보드로부터 학번, 이름, 영어점수, C-언어 점수, 파이썬 점수를 입력받아 총점, 평균, 학점, 등수를 계산하는 프로그램 작성
## 입력 함수, 총점/평균 계산 함수,  학점계산 함수, 등수계산 함수, 출력 함수로 나누어 구현

# 메뉴 - 필요하면 반복문으로 구현
def menu():
    # 5명의 학생 관리
    num = int(input('학생 수: '))

    # 함수 순차 실행
    students = Input(num)     
    students = Cal_Score(students, num)
    
    # 계산된 평균 점수를 기준으로 학점, 순위를 매기기 위해 평균이 저장된 인덱스를 전달
    students = Grade(students, 6)   
    students = Rank(students, 6)
    num_over80 = Count_Over80(students, 6)
    Print_stu(students, num_over80)

# 학생 정보 입력 함수
def Input(num):
    data = []  # 학생 정보를 저장할 리스트
    
    for _ in range(num):
        stu_num = int(input('학번: '))
        name = input('이름: ')
        eng = int(input("영어: "))
        lan_c = int(input("C-언어: "))
        lan_pyt = int(input("파이썬: "))

        # 학생 정보를 리스트로 저장한 후, data에 추가
        student_info = [stu_num, name, eng, lan_c, lan_pyt]
        data.append(student_info)

        print('\n')
    
    # print(data)
    return data  # 입력받은 학생 정보를 반환

# 각 학생의 총점, 평균 계산 함수
def Cal_Score(students, num):
    
    # data의 각 리스트를 가르키는 student에 대해 인덱싱하는 방법
    for student in students:
        total = student[2] + student[3] + student[4]  # 영어 + C-언어 + 파이썬
        average = round(total / 3, 2) # 소수점 2자리까지 저장
        student.append(total)  # 리스트에 총점 추가
        student.append(average) # 리스트에 평균 추가

    return students

# 각 학생의 학점 부여 함수 - 여기까지 완료
def Grade(students, idx):
    for student in students:
        if student[idx] >= 95:
            grade = 'A+'
        elif student[idx] >= 90:
            grade = 'A'
        elif student[idx] >= 85:
            grade = 'B+'
        elif student[idx] >= 80:
            grade = 'B'
        elif student[idx] >= 75:
            grade = 'C+'
        elif student[idx] >= 70:
            grade = 'C'
        elif student[idx] >= 65:
            grade = 'D+'
        elif student[idx] >= 60:
            grade = 'D'
        else:
            grade = 'F'
        student.append(grade)

    # print(students)
    return students

# 각 학생의 등수 부여 함수
def Rank(students, idx):
    # 평균 점수 기준으로 등수 매기기 위해 평균 점수만 추출
    avg_scores = [student[idx] for student in students]

    # 등수를 저장할 리스트 (모든 등수를 1등으로 초기화)
    ranks = [1] * len(students)

    for i in range(len(avg_scores)):
        for j in range(len(avg_scores)):
            # 본인이 점수와 다른 점수들을 모두 비교
            # 자기보다 점수가 높으면 등수 +1
            if avg_scores[i] < avg_scores[j]:  
                ranks[i] += 1  

    # 최종 등수를 students리스트에 추가
    for i in range(len(students)):
        students[i].append(ranks[i])

    return students


def Count_Over80(students, idx):
    count = 0
    for student in students:
        if student[idx] >= 80:
            count += 1

    return count



def Print_stu(students, num80):
    print("\t\t\t\t성적 관리 프로그램\t\t\t\t")
    print("=============================================================================")
    print("학번         이름         영어  C-언어\t파이썬\t총점\t평균\t학점\t등수")
    print("=============================================================================")

    for student in students:
        for i in range(len(student)):
            print(student[i], end="\t")  # 각 데이터를 탭으로 구분하여 출력
        print()

    print(f"\n평균 점수가 80점 이상인 학생 수는 {num80}명입니다.\n")

         
# 시작함수 호출
menu()
```
